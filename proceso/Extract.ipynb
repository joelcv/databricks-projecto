```python

ventas_bronze = spark.read \
    .option("header", True) \
    .csv("abfss://raw@adlsdemodatabricks21.dfs.core.windows.net/detalleVenta.csv")

# Guardar como Delta (Bronze)
ventas_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .save("abfss://bronze@adlsdemodatabricks21.dfs.core.windows.net/ventas_delta")

```

```python
raw_path = "abfss://raw@adlsdemodatabricks21.dfs.core.windows.net/detalleVenta.csv"
bronze_path = "abfss://bronze@adlsdemodatabricks21.dfs.core.windows.net/ventas_delta"

df_raw = spark.read.format("csv") \
    .option("header", True) \
    .option("inferSchema", True) \
    .load(raw_path)

df_raw.write.format("delta") \
    .mode("overwrite") \
    .save(bronze_path)
```
