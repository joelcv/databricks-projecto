```python
from pyspark.sql.functions import col, to_date

bronze_path = "abfss://bronze@adlsdemodatabricks21.dfs.core.windows.net/ventas/"
silver_path = "abfss://silver@adlsdemodatabricks21.dfs.core.windows.net/ventas/"

df = spark.read.format("delta").load(bronze_path)

df_clean = df \
    .withColumn("fecha", to_date(col("fecha"))) \
    .filter(col("monto") > 0)

df_clean.write.format("delta") \
    .mode("overwrite") \
    .save(silver_path)
```
