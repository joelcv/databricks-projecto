```python
ventas = spark.read.format("delta") \
    .load("abfss://silver@adlsdemodatabricks21.dfs.core.windows.net/ventas/")

clientes = spark.read.format("delta") \
    .load("abfss://silver@adlsdemodatabricks21.dfs.core.windows.net/clientes/")

df_gold = ventas.join(clientes, "id_cliente")

df_gold.write.format("delta") \
    .mode("overwrite") \
    .save("abfss://gold@adlsdemodatabricks21.dfs.core.windows.net/ventas_modelo/")
```
