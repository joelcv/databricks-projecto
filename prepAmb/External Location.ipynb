```python
%sql
create widget text storageName default "adlsdemodatabricks21";
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-raw`
URL 'abfss://raw@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación de datasets csv'
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-bronze`
URL 'abfss://bronze@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para tablas bronze'
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-silver`
URL 'abfss://silver@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para tablas silver'
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-golden`
URL 'abfss://golden@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para tablas golden'
```

```python
%sql
CREATE CATALOG IF NOT EXISTS ventas_accesorios
MANAGED LOCATION 'abfss://data@${storageName}.dfs.core.windows.net/'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS ventas_accesorios.bronze;
CREATE SCHEMA IF NOT EXISTS ventas_accesorios.silver;
CREATE SCHEMA IF NOT EXISTS ventas_accesorios.golden;
```
